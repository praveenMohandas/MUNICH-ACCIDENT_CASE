```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# load the dataset
df = pd.read_csv('monatszahlen2209_verkehrsunfaelle.csv')


list(df.columns)
df.replace('Unknown',np.nan, inplace=True)


df.drop(['VORJAHRESWERT','VERÄND_VORMONAT_PROZENT','VERÄND_VORJAHRESMONAT_PROZENT','ZWÖLF_MONATE_MITTELWERT'],axis=1,inplace=True)

#df.columns
#df.dtypes

```

```python
import calendar

# Number of accident per month
accidents_category = df.groupby(df['MONATSZAHL']).count()


accidents_category
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

# plot accidents per category
plot1=accidents_category['WERT']
accidents_category['WERT'].plot(kind='bar',figsize=(12,7), color='red', alpha=0.5)

# title and x,y labels
plt.title('Accidents per category', fontsize=20)
plt.xlabel('Category',fontsize=16)
plt.ylabel('Number of accidents',fontsize=16);
```
